```python
import pandas as pd
import os


df_genders = pd.read_csv(os.path.join("data", "genders.csv"))
df_genders = df_genders.set_index("# user")
df_genders.index.name = "user"
df_genders.head()
```

```python
df_fb = pd.read_csv(os.path.join("data", "fb_friends.csv"))

count_user_a = df_fb.groupby("# user_a").size()
count_user_b = df_fb.groupby("user_b").size()

df_fb_count = count_user_a.add(count_user_b, fill_value=0).to_frame(name="fb_friend_count")

df_fb_count.index.name = "user"
df_fb_count.head(20)
```

```python
df_sms = pd.read_csv(os.path.join("data", "sms.csv"), usecols = ['sender','recipient'])


df_sms_count = pd.concat([df_sms.groupby("sender").size().to_frame(name="sender_count"), df_sms.groupby("recipient").size().to_frame(name="recipient_count")], axis=1)

#Om vi vill ha bara totala antalet sms ist√§llet
#count_sender = df_sms.groupby("sender").size()
#count_recipient = df_sms.groupby("recipient").size()

#df_sms_count = count_sender.add(count_recipient, fill_value=0).to_frame(name="sms_count")

df_sms_count.index.name = "user"
df_sms_count.head()
```

```python
df_calls = pd.read_csv(os.path.join("data", "calls.csv"), usecols = ['caller','callee'])

df_call_count = pd.concat([df_calls.groupby("caller").size().to_frame(name="caller_count"), df_calls.groupby("callee").size().to_frame(name="callee_count")], axis=1)

#Om vi vill ha bara totala antalet samtal ist√§llet
#count_caller = df_calls.groupby("caller").size()
#count_callee = df_calls.groupby("callee").size()

#df_call_count = count_caller.add(count_callee, fill_value=0).to_frame(name="call_count")

df_call_count.index.name = "user"
df_call_count.head(20)


```

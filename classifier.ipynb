```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

df = pd.read_csv("merged.csv")

df.head(10)
```

```python
df_num_feat = df[["fb_friend_count", 
                  "sender_count", 
                  "recipient_count", 
                  "caller_count", 
                  "callee_count"]]

normalized_df_subset=(df_num_feat-df_num_feat.min())/(df_num_feat.max()-df_num_feat.min())

normalized_df_subset.head(10)

df = pd.concat([df[["user", "gender"]], normalized_df_subset, df[["label"]]], axis=1)
df = df.set_index("user")


df = df.dropna()
df.tail(50)
```

```python
X = df.drop("label", axis=1)
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(X, y)

clf = RandomForestClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

accuracy_score(y_test, y_pred)
```
